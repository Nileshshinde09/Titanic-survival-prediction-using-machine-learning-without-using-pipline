```python
le = LabelEncoder()
dataset["Sex"] = le.fit_transform(dataset["Sex"])
```

```python
dataset = dataset.drop(columns = "Ticket",axis = 1)
```

```python
dataset.head()
```

```python
y = dataset.drop(columns = ["Parch","PassengerId","Pclass","Sex","Age","SibSp","Fare","Embarked"], axis = 1)
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=11)
```

```python
regression = LogisticRegression()
```

# Predicting values for train data

```python
regression.fit(x_train,y_train)
```

```python
y_pred = regression.predict(x_train)
```

```python
accuracy_score(y_train,y_pred)*100
```

# Predicting Values for test data

```python
regression.fit(x_test,y_test)
```

```python
y_pred = regression.predict(x_test)
```
